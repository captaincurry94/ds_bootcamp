```python
class TicTacToe():
    import sys
    def __init__(self):
        self.player1 = 'X'
        self.player2 = 'O'
        self.player_name1 = 'Player 1'
        self.player_name2 = 'Player 2'
        self.winnerlist = {'Player 1':0,'Player 2':0}

        self.round_count = 0
        self.current_play = (0,0)
        self.winner = False
        self.current_player = self.player1
        self.current_player_icon = 'X'
        
        self.board = [[(0, 0), (1, 0), (2, 0)],
                       [(0, 1), (1, 1), (2, 1)],
                       [(0, 2), (1, 2), (2, 2)]
                      ] 
        
    def initialize_game(self):
        self.board = [ [(0, 0), (1, 0), (2, 0)],
                       [(0, 1), (1, 1), (2, 1)],
                       [(0, 2), (1, 2), (2, 2)]
                      ]
        self.winner = False
        self.players_turn()
        
    def players_turn(self):
        for b in self.board:
            print(b)
        
        if self.round_count % 2 == 0:
            self.current_player = self.player_name1
            self.current_player_icon = 'X'
        else:
            self.current_player = self.player_name2
            self.current_player_icon = 'O'
        
        self.current_play = input(f'{self.current_player}, what is your next move? Please input your coordinates (x,y) and submit with ENTER')
        self.check_input()
        
    def check_input(self):
        #validation of input
        if len(tuple(self.current_play.replace(',',''))) == 2:
            self.log_play()
        else:
            print('This input does not work. Please enter the coordinates seperated by a comma (x,y)')
            self.players_turn()
    
    def log_play(self):
        horz = int(self.current_play[0])
        vert = int(self.current_play[-1])
        self.board[vert][horz] = ( ' ' + self.current_player_icon + ' ' )
        
        self.winner_check()
    
    def winner_check(self):
        self.winner = False
        
        for i in range(3):
            if self.board[i][0] == self.board[i][1] and self.board[i][0] == self.board[i][2]:
                self.winner = self.current_player
            elif self.board[0][i] == self.board[1][i] and self.board[0][i] == self.board[2][i]:
                self.winner = self.current_player
            elif self.board[0][0] == self.board[1][1] and self.board[0][0] == self.board[2][2]:
                self.winner = self.current_player
            elif self.board[0][2] == self.board[1][1] and self.board[0][2] == self.board[2][0]:
                self.winner = self.current_player
    
        if self.winner != False:
            self.winnerlist[self.current_player] += 1
            print(f'{self.current_player} won the game!')
            for b in self.board:
                print(b)
            
            self.play_again()
            
        else:
            if self.round_count < 9:
                    self.round_count +=1
                    self.players_turn()
                  
                
            else:
                  print('We have a draw!')
                  self.play_again()
                  
    
    def play_again(self):
        print('Current stats: ' + str(self.winnerlist))
        new_round = input('Want to play another round? Press any key to continue or ‘n’ to end the game')

        if new_round.lower() == 'n':
            sys.exit()

        else: 
            print('NEW GAME')
            self.play_game()
                  
    
    def play_game(self):
        self.initialize_game()
                  
```
